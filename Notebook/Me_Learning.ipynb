```python
%matplotlib inline
import matplotlib.pyplot as plt

x = np.arange(0,3*np.pi,0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.plot(x,y_sin)
plt.plot(x,y_cos)
plt.xlabel("Independent Variable")
plt.ylabel("Dependent Variable")
plt.legend(["Sine","Cosine"])
plt.title("Sine and Cosine Functions")
plt.show()
```

```python
plt.subplot(2,1,1)
plt.plot(x,y_sin)
plt.title("Sine")

plt.subplot(2,1,2)
plt.plot(x,y_cos)
plt.title("Cosine")

plt.show()
```
