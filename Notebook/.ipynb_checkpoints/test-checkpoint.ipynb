```python
# Add Ceiling

fare_ceiling = 40
data[ data[:,9].astype(np.float)> fare_ceiling, 9 ] = fare_ceiling -1
print data[:,9]
```

```python
fare_bracket = 10
num_brackets = fare_ceiling/fare_bracket
print num_brackets
```

```python
#num_classes = 3
num_classes =  len(np.unique(data[:,2]))
print num_classes

num_genders = len(np.unique(data[:,4]))
print num_genders
```

```python
survival_table = np.zeros((num_genders,num_classes,num_brackets))
print survival_table
```

```python
for i in xrange(num_classes):
    for j in xrange(num_brackets):
        
        bool_idx_w = (data[:,4] == "female") and (data[:,2].astype(np.float)=i+1) \
                    and (data[:,9].astype(np.float)>=j*fare_bracket) \
                    and (data[:,9].astype(np.float)<(j+1)*fare_bracket)
                
        bool_idx_m = (data[:,4] == "male") and (data[:,2].astype(np.float)=i+1) \
                    and (data[:,9].astype(np.float)>=j*fare_bracket) \
                    and (data[:,9].astype(np.float)<(j+1)*fare_bracket)
                
        survival_table[0,i,j] = np.mean(data[bool_idx_w,2])
        survival_table[1,i,j] = np.mean(data[bool_idx_m,2])
```
